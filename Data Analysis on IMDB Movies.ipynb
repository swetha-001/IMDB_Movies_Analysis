Classify Movies based on Ratings(Excellent, Good and Average)

```python
# create a function to classify
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['Rating_Category']=data['Rating'].apply(rating)
```

Find Unique Values From Genre

```python
data['Genre']
```

To get unique values :
1. Split the values seperated by ',' into a 2D list
2. convert the 2D list into 1D list
3. find unique values from 1D list

```python
list_2D=[]
for item in data['Genre']:
    list_2D.append(item.split(','))
```

```python
list_2D
```

```python
list_1D=[]
for item1 in list_2D:
    for item2 in item1:
        list_1D.append(item2)
```

```python
list_1D
```

```python
unique_Genres=[]
for item in list_1D:
    if item not in unique_Genres:
        unique_Genres.append(item)
```

```python
unique_Genres
```
